# Model Training

```python
#splitting the data into input and out variable
```

```python
x=df.drop(columns=['ISO3'],axis=1)
```

```python
x.head()
```

```python
y=df.ISO3
```

```python
y.head()
```

```python
#checking the shape of input and target variable
```

```python
print(x.shape,'\t\t',y.shape)
```

```python
#Scaling in input variable
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x=ss.fit_transform(x)
```

```python
#splitting into training and testing data
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.20,random_state=42)
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
#importing all the model Libreary
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lg=LogisticRegression()
lg.fit(x_train,y_train)
```

```python
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

#importing Boosting Models

from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import ExtraTreesClassifier

#import error metrics

from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
from sklearn.model_selection import GridSearchCV,cross_val_score
```

```python
#importing all the algoritham
```

```python
model=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier(),BaggingClassifier(),ExtraTreesClassifier()]

for m in model:
    m.fit(x_train,y_train)
    m.score(x_train,y_train)
    predm=m.predict(x_test)
    print('Accuracy Score of',m,'is:')
    print(accuracy_score(y_test,predm))
    print(confusion_matrix(y_test,predm))
    print(classification_report(y_test,predm))
    print('*********************************************************')
    print('\n')
```

All the models are performing equaly better

```python
#Cross validate the model
```

```python
model=[LogisticRegression(),GaussianNB(),SVC(),DecisionTreeClassifier(),KNeighborsClassifier(),RandomForestClassifier(),
      AdaBoostClassifier(),GradientBoostingClassifier(),BaggingClassifier(),ExtraTreesClassifier()]

for m in model:
    score=cross_val_score(m,x,y,cv=15,scoring='accuracy')
    print("model:",m)
    print("score:",score)
    print("mean score:",score.mean())
    print("standard deviation:",score.std())
    print('*****************************')
    print('\n')
    
```

Accept Decision Tree Classifier rest all models performed better

```python
#find the best parameters
```

using GridSearchCV to find out the best parameters in Logistic regression

```python
from sklearn.model_selection import GridSearchCV
```

```python
parameters={'C':[1,10],'random_state':range(42,100)}
lg=LogisticRegression()
clf=GridSearchCV(lg,parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
#LogisticRegrssion with Best parameters
```

```python
print(accuracy_score(y_test,predlg))
print(confusion_matrix(y_test,predlg))
print(classification_report(y_test,predlg))
```

Using gridsearchcv to find out the best parameter in svm

```python
parameters={'kernel':('linear','rbf','poly','sigmoid'),'C':[1,10],'random_state':range(42,60)}
sv=SVC()
clf=GridSearchCV(sv,parameters)
clf.fit(x,y)
print(clf.best_params_)
```

```python
sv=SVC(kernel='linear',C=1,random_state=42,probability=True)
sv.fit(x_train,y_train)
sv.score(x_train,y_train)
predsv=sv.predict(x_test)
print(accuracy_score(y_test,predsv))
print(confusion_matrix(y_test,predsv))
print(classification_report(y_test,predsv))
```

```python
#cross validate SVC
```

```python
score=cross_val_score(sv,x,y,cv=5,scoring='accuracy')
print(score)
print('mean:',score.mean())
print('std:',score.std())
```
